```python
import random
import re
flag=False 
n=0                                 # counter for trials
while(flag==False):
    t=input("choose easy or average or hard:")
    if t == 'easy':
        x=random.randrange(0,10)
        if flag==False:                      #no win yet
            while(True):      
                z=None
                z=input("guess a number from 0 to 10:")    # takes input from user
                if re.match('\d',z):          #check if the input is a number
                    z=int(z)     
                    if 0 <= z <= 10:          #check if the input is in the range
                        if z == x:
                            print ("you guessed right")
                            flag=True         # he guessed right and get out loop
        
                            break
                        if z > x:
                            n+=1              # one trial passed and the number is lower
                            print("wrong number your number is higher")
                        if z < x:             # one trial passed and the number is higher
                            print("wrong number your number is lower")
                            n+=1
                        if n == 3:             #trials ended
                            print("you lost the number was",x)
                    else:
                        print("you entered number out of range")
                else:
                    print("you didnt enter a number")
    if t == 'average':
        x=random.randrange(0,50)
        if flag==False:                      #no win yet
            while(True):      
                z=None
                z=input("guess a number from 0 to 50:")    # takes input from user
                if re.match('\d',z):          #check if the input is a number
                    z=int(z)     
                    if 0 <= z <= 50:          #check if the input is in the range
                        if z == x:
                            print ("you guessed right")
                            flag=True         # he guessed right and get out loop
        
                            break
                        if z > x:
                            n+=1              # one trial passed and the number is lower
                            print("wrong number your number is higher")
                        if z < x:             # one trial passed and the number is higher
                            print("wrong number your number is lower")
                            n+=1
                        if n == 3:             #trials ended
                            print("you lost the number was",x)
                    else:
                        print("you entered number out of range")
                else:
                    print("you didnt enter a number")
    if t == 'hard':
        x=random.randrange(0,100)
        if flag==False:                      #no win yet
            while(True):      
                z=None
                z=input("guess a number from 0 to 100:")    # takes input from user
                if re.match('\d',z):          #check if the input is a number
                    z=int(z)     
                    if 0 <= z <= 100:          #check if the input is in the range
                        if z == x:
                            print ("you guessed right")
                            flag=True         # he guessed right and get out loop
        
                            break
                        if z > x:
                            n+=1              # one trial passed and the number is lower
                            print("wrong number your number is higher")
                        if z < x:             # one trial passed and the number is higher
                            print("wrong number your number is lower")
                            n+=1
                        if n == 3:             #trials ended
                            print("you lost the number was",x)
                    else:
                        print("you entered number out of range")
                else:
                    print("you didnt enter a number")
    if (t!="easy") or (t!= "hard") or (t!= "average"):
        print("you can only choose easy or average or hard")
```
